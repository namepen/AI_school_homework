```python
class Naive_Bayes_Classifier(object):
    # 48 feature 
    def __init__(self, num_features=48):
        self.num_features = num_features

    # (2) log(P(feature_vector | Class)) .
    def log_likelihood_naivebayes(self, feature_vector, Class): #feature_vector   log_likehood 
        assert len(feature_vector) == self.num_features
        log_likelihood = 0.0 #log-likelihood  underflow 
        
        if Class == 0:
            zero_sum = np.sum(np.sum(X_train[i] for i in range(len(X_train)) if y_train[i] == 0))

            for j in range(num_features):
                log_likelihood = log_likelihood * np.sum(X_train[i][j] for i in range(len(X_train)) if y_train[i] == 0) 
                                        / zero_sum
        elif Class == 1:
            one_sum = np.sum(np.sum(X_train[i] for i in range(len(X_train)) if y_train[i] == 1))

            for j in range(num_features):
                log_likelihood = log_likelihood * np.sum(X_train[i][j] for i in range(len(X_train)) if y_train[i] == 1) 
                                        / zero_sum
    
        else:
            raise ValueError("Class takes integer values 0 or 1")
        
        return log_likelihood
    
    # 3.   Posterior probability .
    def class_posteriors(self, feature_vector):
        log_likelihood_ham = self.log_likelihood_naivebayes(feature_vector, Class = 0)
        log_likelihood_spam = self.log_likelihood_naivebayes(feature_vector, Class = 1)
        
        log_posterior_ham = log_likelihood_ham * 0.5
        log_posterior_spam = log_likelihood_spam * 0.5
        
        return log_posterior_ham, log_posterior_spam
    
    # Maximum A Priori(MAP) inference      .
    def spam_classify(self, document):
        feature_vector = [int(element>0.0) for element in document]
        log_posterior_ham, log_posterior_spam = self.class_posteriors(feature_vector)
        
        if log_posterior_ham > log_posterior_spam:
            return 0
        else:
            return 1


    # (1)  train  .
    def train(self, X_train, y_train):
        # Likelihood estimator 
        #     
        X_train_spam = [X_train[i] for i in range(len(X_train)) if y_train[i] == 1]
        X_train_ham = [X_train[i] for i in range(len(X_train)) if y_train[i] == 0]
    
        #   feature   likelihood .
        self.likelihoods_ham = np.mean(X_train_ham, axis = 0)/100.0
        self.likelihoods_spam = np.mean(X_train_spam, axis = 0)/100.0

        #  class log-prior 
        num_ham = float(len(X_train_ham))
        num_spam = float(len(X_train_spam))

        prior_probability_ham = num_ham / float(len(X_train))
        prior_probability_spam = num_spam / float(len(X_train))

        self.log_prior_ham = np.log10(prior_probability_ham)
        self.log_prior_spam = np.log10(prior_probability_spam)
    
        return np.argmax([log_prior_ham, log_prior_spam])
    
    def predict(self, X_test):
        predictions = []
        for case in X_test:
            predictions.append(self.spam_classify(case))
        
        return predictions
```

```python
NB = Naive_Bayes_Classifier()
NB.train(X_train, y_train)
pred = NB.predict(X_test)

def evaluate_performance(predictions, ground_truth_labels):
    correct_count = 0.0
    for item_index in range(len(predictions)):
        if predictions[item_index] == ground_truth_labels[item_index]:
            correct_count += 1.0
    accuracy = correct_count/len(predictions)
    return accuracy

accuracy_naivebayes = evaluate_performance(pred, y_test)
print(accuracy_naivebayes)
```

```python
# Gaussian Naive Bayes
import numpy as np
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#   
dataset = datasets.load_breast_cancer()

#  
X_train, X_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size = 0.25, random_state=21)
```

```python
# (1) Gaussian Naive Bayes    fit .
model_gaussianNB = GaussianNB()
model_gaussianNB.fit(X=X_train, y=y_train)
print(model_gaussianNB)
```

```python
# (2) Logistic Regression    fit .
model_logisticReg = LogisticRegression()
model_logisticReg.fit(X=X_train, y=y_train)
print(model_logisticReg)
```

```python
# (3) Fitting    test  label predict .
expected = y_test
pred_gaussianNB = model_gaussianNB.predict(X=X_test)
pred_logistcReg = model_logisticReg.predict(X=X_test)
```

```python
# Prediction 
print(metrics.classification_report(expected, pred_gaussianNB))
print(metrics.classification_report(expected, pred_logistcReg))
```
