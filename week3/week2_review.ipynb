#### Machine Learning overview

Learning이란?

 - 대부분의 경우에 data는 싸고 풍부하다(abundant). 하지만, 정보은 비싸고 귀하다
 - 컴퓨터를 이용해서 데이터를 분석(analyze)하고 정보를 추출하는 모델을 만드는 것을 Learning이라고 한다.
 - 목적 : 특정한 샘플 데이터(particular exmaples)에서 일반적인 룰(general rules)를 찾는 과정
 - 데이터에 적합한 모델을 구축해야 한다.

#### Feature Extraction/representation

- 원래의 input variable을 새로운 공간의 variable로 변환하는 과정
- 새로운 공간(new space)에서 문제를 쉽게 해결할 수 있도록 구성
- 새로운 데이터는 training data로 사용하기 위해 전처리 과정(pre-processing)을 거친 후 사용한다

#### Terminology

1. Supervised learning(지도 학습, 답이 정해져 있는 data 학습)
  - Regression(회귀) : continuous outputs
  - classfication(분류) : discrete(이산) or category outputs
  
  
  
2. Unsupervised learning(비지도 학습, 답이 없는 data 학습)
  - clustering
  - Density estimation
  - Visualization
  
  
3. Reinforcement learning
  - 보상을 최대화하기 위한 적합한 액션(suitable action)을 찾는 과정

#### Linear Classification Models

- Linear separable: Data set의 class를 선형 결정 평면(linear decision surface)로 나누는 과정

- class Label representations
    1. Two class(K=2) : binary representation
    2. K>2 : 1-of-K coding scheme

#### Different Approaches to classification

1. Discriminant function
    - X를 특정한 class로 바로 할당한다(directly assign)
    - 예시) Fisher's linear discriminant, perceptron

2. Probabilistic models
